```python
import pywt
import math 
import numpy as np
```

```python
db1 = pywt.Wavelet('haar')
```

```python
xi = [100,101,102,103,104,105,106,107,108,109,111,112,113,114,115,116,117,118,119,121,122,123,124,200,201,202,203,205,207,208,209,210,212,213,214,215,217,219,220,221,222,223,228,230,231,232,233,234]
```

```python
def prep():
    a = []
    for i in xi:
        record ,field= wfdb.rdsamp(f"../../../input/ANFIS/physionet.org/files/mitdb/1.0.0/{i}", sampto=3000)
        tx1,tx2 = np.array(pywt.dwt(record[:,0], db1))
        v1=np.append(tx1,tx2)
        a.append(np.array(v1))
    return a
```

```python
X=pd.DataFrame(prep())
```

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
```

```python
kmeans = KMeans(n_clusters=6, random_state=0).fit(X)
```

```python
y=kmeans.labels_
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```
