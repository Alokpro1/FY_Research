```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
def home_eval(mkt,loc):
    market_value = ctrl.Antecedent(np.arange(0, 101, 1), 'market value')
    location = ctrl.Antecedent(np.arange(0, 101, 1), 'location')
    house = ctrl.Consequent(np.arange(0, 101, 1), 'house')

    market_value['low'] = fuzz.trapmf(market_value.universe, [0, 0, 5, 10])
    market_value['medium'] = fuzz.trapmf(market_value.universe, [5, 10,20, 25])
    market_value['high'] = fuzz.trapmf(market_value.universe, [20, 25, 65,85])
    market_value['very high'] = fuzz.trapmf(market_value.universe, [65, 85,100, 100])

    location['bad'] = fuzz.trapmf(location.universe, [0, 0, 15, 40])
    location['fair'] = fuzz.trapmf(location.universe, [25, 50, 60, 85])
    location['excellent'] = fuzz.trapmf(location.universe, [60, 85, 85,100])

    house['very low'] = fuzz.trimf(house.universe, [0, 0, 30])
    house['low'] = fuzz.trimf(house.universe, [0, 30, 60])
    house['medium'] = fuzz.trimf(house.universe, [20, 50, 80])
    house['high'] = fuzz.trimf(house.universe, [40, 70, 100])
    house['very high'] = fuzz.trimf(house.universe, [ 70,100, 100])
    
#     market_value.view()
#     location.view()
#     house.view()
    
    rule1 = ctrl.Rule(market_value['low'] , house['low'])
    rule2 = ctrl.Rule(location['bad'], house['low'])
    rule3 = ctrl.Rule(market_value['low'] & location['bad'], house['very low'])
    rule4 = ctrl.Rule(market_value['medium'] & location['bad'], house['low'])
    rule5 = ctrl.Rule(market_value['high'] & location['bad'], house['medium'])
    rule6 = ctrl.Rule(market_value['very high'] & location['bad'], house['high'])
    rule7 = ctrl.Rule(market_value['low'] & location['fair'], house['low'])
    rule8 = ctrl.Rule(market_value['medium'] & location['fair'], house['medium'])
    rule9 = ctrl.Rule(market_value['high'] & location['fair'], house['high'])
    rule10 = ctrl.Rule(market_value['very high'] & location['fair'], house['very high'])
    rule11 = ctrl.Rule(market_value['low'] & location['excellent'], house['medium'])
    rule12 = ctrl.Rule(market_value['medium'] & location['excellent'], house['high'])
    rule13 = ctrl.Rule(market_value['high'] & location['excellent'], house['very high'])
    rule14 = ctrl.Rule(market_value['very high'] & location['excellent'], house['very high'])

    # rule1.view()
    evaluation_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14])
    evaluation = ctrl.ControlSystemSimulation(evaluation_ctrl)
    
    evaluation.input['market value'] = mkt
    evaluation.input['location'] = loc

    evaluation.compute()
    
    print(evaluation.output['house'])
    house.view(sim=evaluation)
```

```python
home_eval(41,25)
```

```python
def app_eval(ast,inc):
    asset = ctrl.Antecedent(np.arange(0, 100, 1), 'asset')
    income = ctrl.Antecedent(np.arange(0, 100, 1), 'income')
    applicant = ctrl.Consequent(np.arange(0, 100, 1), 'applicant')
    
    asset['low'] = fuzz.trimf(asset.universe, [0, 0, 15])
    asset['medium'] = fuzz.trapmf(asset.universe, [5, 25, 45, 65])
    asset['high'] = fuzz.trapmf(asset.universe, [50, 70, 100, 100])
    
    income['low'] = fuzz.trapmf(income.universe, [0, 0, 10, 25])
    income['medium'] = fuzz.trimf(income.universe, [15, 35, 55])
    income['high'] = fuzz.trimf(income.universe, [40, 60, 80])
    income['very high'] = fuzz.trapmf(income.universe, [60, 80,100, 100])
    
    applicant['low'] = fuzz.trapmf(applicant.universe, [0, 0, 20, 40])
    applicant['medium'] = fuzz.trimf(applicant.universe, [20, 50, 80])
    applicant['high'] = fuzz.trapmf(applicant.universe, [60, 80, 100,100])
    
#     asset.view()
#     income.view()
#     applicant.view()
    
    rule1 = ctrl.Rule(asset['low'] & income['low'], applicant['low'])
    rule2 = ctrl.Rule(asset['low'] & income['medium'], applicant['low'])
    rule3 = ctrl.Rule(asset['low'] & income['high'], applicant['medium'])
    rule4 = ctrl.Rule(asset['low'] & income['very high'], applicant['high'])
    rule5 = ctrl.Rule(asset['medium'] & income['low'], applicant['low'])
    rule6 = ctrl.Rule(asset['medium'] & income['medium'], applicant['medium'])
    rule7 = ctrl.Rule(asset['medium'] & income['high'], applicant['high'])
    rule8 = ctrl.Rule(asset['medium'] & income['very high'], applicant['high'])
    rule9 = ctrl.Rule(asset['high'] & income['low'], applicant['medium'])
    rule10 = ctrl.Rule(asset['high'] & income['medium'], applicant['medium'])
    rule11 = ctrl.Rule(asset['high'] & income['high'], applicant['high'])
    rule12 = ctrl.Rule(asset['high'] & income['very high'], applicant['high'])
    
    evaluation_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12])
    evaluation = ctrl.ControlSystemSimulation(evaluation_ctrl)
    
    evaluation.input['asset'] = ast
    evaluation.input['income'] = inc

    evaluation.compute()
    
    print(evaluation.output['applicant'])
    applicant.view(sim=evaluation)
```

```python
def credit_eval(inc,intr,app,hus):
    income = ctrl.Antecedent(np.arange(0, 100, 1), 'income')
    interest = ctrl.Antecedent(np.arange(0, 100, 1), 'interest')
    applicant = ctrl.Antecedent(np.arange(0, 100, 1), 'applicant')
    house = ctrl.Antecedent(np.arange(0, 100, 1), 'house')
    credit = ctrl.Consequent(np.arange(0, 100, 1), 'credit')
    
    income['low'] = fuzz.trapmf(income.universe, [0, 0, 10, 25])
    income['medium'] = fuzz.trimf(income.universe, [15, 35, 55])
    income['high'] = fuzz.trimf(income.universe, [40, 60, 80])
    income['very high'] = fuzz.trapmf(income.universe, [60, 80,100, 100])
    
    interest['low'] = fuzz.trapmf(interest.universe, [0, 0, 20, 50])
    interest['medium'] = fuzz.trapmf(interest.universe, [20, 40, 60, 80])
    interest['high'] = fuzz.trapmf(interest.universe, [60, 85, 100,100])
    
    applicant['low'] = fuzz.trapmf(applicant.universe, [0, 0, 20, 40])
    applicant['medium'] = fuzz.trimf(applicant.universe, [20, 50, 80])
    applicant['high'] = fuzz.trapmf(applicant.universe, [60, 80, 100,100])
    
    house['very low'] = fuzz.trimf(house.universe, [0, 0, 30])
    house['low'] = fuzz.trimf(house.universe, [0, 30, 60])
    house['medium'] = fuzz.trimf(house.universe, [20, 50, 80])
    house['high'] = fuzz.trimf(house.universe, [40, 70, 100])
    house['very high'] = fuzz.trimf(house.universe, [ 70,100, 100])
    
    credit['very low'] = fuzz.trimf(credit.universe, [0, 0, 30])
    credit['low'] = fuzz.trimf(credit.universe, [0, 30, 60])
    credit['medium'] = fuzz.trimf(credit.universe, [20, 50, 80])
    credit['high'] = fuzz.trimf(credit.universe, [40, 70, 100])
    credit['very high'] = fuzz.trimf(credit.universe, [ 70,100, 100])
    
#     income.view()
#     interest.view()
#     applicant.view()
#     house.view()
#     credit.view()
    
    rule1 = ctrl.Rule(income['low'] & interest['medium'], credit['low'])
    rule2 = ctrl.Rule(income['low'] & interest['high'], credit['very low'])
    rule3 = ctrl.Rule(income['medium'] & interest['high'], credit['low'])
    rule4 = ctrl.Rule(applicant['low'], credit['very low'])
    rule5 = ctrl.Rule(house['very low'], credit['very low'])
    rule6 = ctrl.Rule(applicant['medium'] & house['very low'], credit['low'])
    rule7 = ctrl.Rule(applicant['medium'] & house['low'], credit['low'])
    rule8 = ctrl.Rule(applicant['medium'] & house['medium'], credit['medium'])
    rule9 = ctrl.Rule(applicant['medium'] & house['high'], credit['high'])
    rule10 = ctrl.Rule(applicant['medium'] & house['very high'], credit['high'])
    rule11 = ctrl.Rule(applicant['high'] & house['very low'], credit['low'])
    rule12 = ctrl.Rule(applicant['high'] & house['low'], credit['medium'])
    rule13 = ctrl.Rule(applicant['high'] & house['medium'], credit['high'])
    rule14 = ctrl.Rule(applicant['high'] & house['high'], credit['high'])
    rule15 = ctrl.Rule(applicant['high'] & house['high'], credit['very high'])
    
    evaluation_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15])
    evaluation = ctrl.ControlSystemSimulation(evaluation_ctrl)
    
    evaluation.input['income'] = inc
    evaluation.input['interest'] = intr
    evaluation.input['applicant'] = app
    evaluation.input['house'] = hus

    evaluation.compute()
    
    print(evaluation.output['credit'])
    credit.view(sim=evaluation)
    
    
    
    
    
    
    
    
```
